## Group By

Any groupby operation involves one of the following operations on the original object. They are âˆ’

- Splitting the Object

- Applying a function

- Combining the results

In many situations, we split the data into sets and we apply some functionality on each subset. In the apply functionality, we can perform the following operations âˆ’

- **Aggregation** âˆ’ computing a summary statistic

- **Transformation** âˆ’ perform some group-specific operation

- **Filtration** âˆ’ discarding the data with some condition

````python
# group imdb_df by movie genres
imdb_df.groupby(['genre'])
````

````python
# get crime movies group
group=imdb_df.groupby(['genre'])
group.get_group('Crime')
````

````python
# get mean of movie durations for each group
# imdb_df['duration'].mean()
genre=imdb_df.groupby(['genre'])
meanDuration=genre['duration'].mean()
meanDuration
````

````python
# change duration of all movies in a particular genre to mean duration of the group
size=len(imdb_df)
for i in range(1,size):
    imdb_df.loc[i,'duration']=meanDuration[imdb_df.loc[i,'genre']]
imdb_df
    
    
````

````python
# drop groups/genres that do not have average movie duration greater than 120.
for i in range(len(imdb_df)):
    if(not imdb_df.loc[i,'duration']>120):
        imdb_df=imdb_df.drop(i)
imdb_df        
        
````

````python
# group weekday wise bikers count
group_weekday=bikes_df.groupby(['weekday'])
group_weekday
bikes_df
````

````python
# get weekday wise biker count
weekday=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
for day in weekday:
    weekday_count=group_weekday.get_group(day).sum(axis=1)
    day_count=weekday_count.sum()
    print(day,":",day_count)
````

````python
# plot weekday wise biker count for 'Berri1'
li=[]
for day in weekday:
    Berri1_count=group_weekday.get_group(day).sum()
    li.append(Berri1_count['Berri1'])
li    
plt.plot(li)    
    
````
